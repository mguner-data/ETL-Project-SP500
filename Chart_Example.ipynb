```python
import pandas as pd
from sqlalchemy import create_engine
from secret import username, password
import matplotlib.pyplot as plt
```

```python
connection_string = f'{username}:{password}@localhost:5432/Stock_DB'
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
df= pd.read_sql_table('daily_data',engine)
```

```python
ticker1='AAPL'
ticker2='MSFT'
dfa = df[df['ticker']==ticker1]
dfm = df[df['ticker']==ticker2]
plt.figure(figsize=(20,10))
plt.rcParams.update({'font.size': 22})
plt.plot(dfa['date_close'],dfa['close'],label=ticker1)
plt.plot(dfm['date_close'],dfm['close'],label=ticker2)
plt.legend()
plt.show
```
