### Short Code Used to List All Unique Symbols to Cover Relevant Years

```python
import datetime as dt
import numpy as np
import pandas as pd
```

```python
# Read csv files
#1) From data sources
df2=pd.read_csv("constituents.csv")
#From all symbols
symbols_all=pd.read_csv("Symbols.csv")
symbols_all.info()
```

```python
# Get all ticker symbols 
list(symbols_all.columns.values)
ticker_list=symbols_all['Symbol'].to_list()
set_ticker_list=set(ticker_list)
ticker_lst=list(set_ticker_list)
ticker_lst
```

```python
ticker_series=pd.Series(ticker_lst, name="Symbol")

df_full_tickers=ticker_series.to_frame()
merged=df_full_tickers.merge(df2, on="Symbol", how="outer")

merged.to_csv('unique_sym.csv', index=False)
```
